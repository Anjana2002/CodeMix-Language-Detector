```python
import re
import pandas as pd
import contractions
```

```python
df = pd.read_csv("malayalam_youtube_comments.csv")
df.head()
```

```python
#check row contain malayalam language
malayalam_pattern = r'[\u0D00-\u0D7F]+'
def remove_malayalam(text):
    if isinstance(text, str):
        return re.sub(malayalam_pattern, '', text)
    return text
df = df.applymap(remove_malayalam)
```

```python
df.head()
```

```python
df = df.applymap(lambda x: ' '.join(str(x).split()) if isinstance(x, str) else x)
df = df.replace(r'^\s*$', None, regex=True)
df = df.dropna(how='any')
```

```python
print(df.isnull().sum())
print(df.info())
```

```python
df = df[df['text'].apply(lambda x: len(x.split()) >= 2 if isinstance(x, str) else False)]
```

```python
def trim_repeated(word):
  return re.sub(r'(.)\1{2,}',r'\1\1', word)
df['text'] = df['text'].apply(trim_repeated)
```

```python
def expand_contractions(text):
    if isinstance(text, str):
        return contractions.fix(text)
    return text
df['text'] = df['text'].apply(expand_contractions)
```

```python
df.to_csv('malayalam_youtube_comments_filtered.csv', index=False)
```
